#String and Sets

```python
#Q1. Write a program to count and display the number of capital letters in a given string.

s = input("Enter string: ")

capitals = set("ABCDEFGHIJKLMNOPQRSTUVWXYZ")
count = 0

for ch in s:
    if ch in capitals:
        count += 1

print("Capital letters:", count)
```

```python
#2. Count total number of vowels in a string.

s = input("Enter a string :")
c = 0
for i in s:
  if i in "aeiouAEIOU":
    print(i)
    c += 1
print(c)
```

```python
#3. Input a sentence and print words in separate lines

s = input("Enter a string: ")
word = s.split()

for i in word:
  print(i)
```

```python
#4. Write a string to check whether the substring exists in the string
s=input("Enter a string: ")
r=input("Enter a substring: ")
count = 0
if r in s:
    print("String exists:")
    count += 1
else:
    print("String doesn't exists:")
print(count)



```

```python
#5. Write a Python program to count the number of occurrences of each alphabet
s = input().upper()

for ch in set(s):
    if ch.isalpha():
        print(s.count(ch), ch, sep="")
```

```python
#6.Program to count number of unique words in a given sentence using sets.
sentence = input("Enter sentence: ").lower()

words = sentence.split()
unique_words = set(words)

print("Number of unique words:", len(unique_words))
```

```python
#7.Create 2 sets s1 and s2 of n fruits each by taking input from user and find:
#a) Fruits which are in both sets s1 and s2
#b) Fruits only in s1 but not in s2
#c) Count of all fruits from s1 and s2

n = int(input("Enter number of fruits in each set: "))

s1 = set()
s2 = set()

print("Enter fruits for Set 1:")
for i in range(n):
    s1.add(input())

print("Enter fruits for Set 2:")
for i in range(n):
    s2.add(input())

print("Common fruits:", s1 & s2)
print("Only in s1:", s1 - s2)
print("Total unique fruits:", len(s1 | s2))
```

```python
#8. Take two sets and apply various set operations on them :
s1 = {"red", "blue", "yellow", "pink"}
s2 = {"pink", "black", "white"}

print(s1 | s2) #union
print(s1 & s2)
print(s1 - s2)
print(s1 ^ s2)
```
